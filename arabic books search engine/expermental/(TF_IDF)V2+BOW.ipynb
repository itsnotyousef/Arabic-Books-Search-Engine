```python
import re
import string
import nltk
from nltk.corpus import stopwords
from snowballstemmer import stemmer
```

```python
nltk.download('stopwords')
arabic_stemmer = stemmer("arabic")
```

```python
def remove_punctuation(text):
    arabic_punctuation = r"[،؛؟!٪ـ«»…\"\'٫٬٭\,.:؛؟!<>*”…“()\[\{}/\\_#$%&+;=?@^~`\]\|➖\u2014\u2013\u002D]"
    return re.sub(arabic_punctuation,' ', text)


def remove_repeated_whitespaces(text):
    return re.sub(r'(?<=\S)\s{2,}(?=\S)', ' ', text)

def remove_arabic_numbers(text):
    return re.sub(r'[\u0660-\u0669]', "", text)

def remove_diacritics(text):
    arabic_diacritics = re.compile(r'[\u064B-\u065F]')
    return re.sub(arabic_diacritics, ' ', text)

def remove_all_except_arabic(text):
    return re.sub(r'[^\u0600-\u06FF\s]', ' ', text)

def remove_links(text):
  return re.sub(r'http[s]?://\S+|www\.\S+',' ',text)

def remove_stop_words(text):
    stop_words = set(stopwords.words('arabic'))  # Corrected function call
    words = text.split()  # Tokenize the text by spaces
    filtered_text = " ".join([word for word in words if word not in stop_words])
    return filtered_text

def normalize_arabic(text):
    # Define normalization mapping
    normalization_map = {
        "إ": "ا",
        "أ": "ا",
        "آ": "ا",
        "ا": "ا",  # Normalize Alef forms
        "ى": "ي",
        "ي": "ي",                      # Normalize Yaa
        "ؤ": "و",
        "ئ": "ي",                      # Normalize Hamza forms
        "ة": "ه",                                # Convert Teh Marbuta to Heh
        "گ": "ك",
        "ڤ": "ف",
        "چ": "ج",
        "پ": "ب",  # Persian/Urdu to Arabic conversion
                       # Convert Arabic numbers to Western numbers
    }

    # Apply replacements using regex
    text = re.sub("|".join(map(re.escape, normalization_map.keys())),
                  lambda m: normalization_map[m.group()], text)

    return text

```

```python
def preprocess(series):
  series = series.apply(remove_diacritics)
  series = series.apply(remove_links)
  series = series.apply(remove_punctuation)
  series = series.apply(remove_arabic_numbers)
  series = series.apply(remove_all_except_arabic)
  series = series.apply(remove_stop_words)
  series = series.apply(remove_repeated_whitespaces)
  series = series.apply(normalize_arabic)
  return series
```

```python
queries = ['''قصة بوليسيه وهي ختام سلسلة الرجل المستحيلهذه السلسلة من أشهر السلاسل البوليسية للشباب.. رجل المستحيل دائمًا ما تجده في قلب الأحداث وقادرًا أيضًا على قلب ذمة الأحداث رأسًا على عقب.. يفكر ويخطط وينفذ ما لم يخطر على بالك قط.. متمرس في أنواع كثيرة من فنون الحياة.. كقتال.. التخطيط.. اتقان اللغات وغيرها كثيرًا.. له عقل ذو إمكانيات فائقة المهارة ولكنه أيضًا له قلب يحب ويشعر. فهو لم يكن أبدًا آلة خالية من المشاعر الإنسانية ولهذا نجد دائمًا لمغامراته طعم ومذاق خاص بها وحدها تلك هي سلسلة رجل المواقف الصعبة بل المستحيلة إنه رجل المستحيل.\nقوات أربع دول ومخابراتها، تطارد رجلًا واحد بلا رحمة.. أكبر منظمتين إجراميتين تنضمان إلى ذلك التحالف الرهيب.. الإسرائيليون يسعون للسيطرة على العالم.. وعلى (أدهم صبري) أن يواجه كل هذا وحده.. وأن يخوض معركة ضد العالم كله لإنقاذ العالم كله.. ترى.. هل ينتصر (أدهم) في معركته الأخيرة، أم إنها مهمة (الوداع)؟! اقرأ التفاصيل المثيرة، وقاتل مع رجل المستحيل في معركته الأخيرة.. بحق.''']
```

```python
query_vector_list = []
for query in queries:
  query_tdidf = vectorizer.transform([query])
  query_vector_list.append(query_tdidf.toarray()[0])
```

```python
import torch
from sentence_transformers import util
import numpy as np

# Convert book embeddings to tensor
book_embeddings_tensor = torch.from_numpy(transformed_data.toarray())

# List to store average similarity scores for each query
similarity_scores = []

# Iterate over each query vector in the list
for query_idx, query_vector in enumerate(query_vector_list):
    query_embedding_tensor = torch.from_numpy(query_vector)

    # Compute cosine similarity (shape: [1, num_books])
    cosine_scores = util.cos_sim(query_embedding_tensor, book_embeddings_tensor)

    # Convert to NumPy array and flatten
    cosine_scores_np = cosine_scores.numpy().flatten()

    # Get top 10 indices sorted by similarity score (descending order)
    top_indices = np.argsort(cosine_scores_np)[::-1][:10]

    # Print the query number
    print(f"\n=== Query {query_idx + 1} ===")

    # Compute and print the top 10 books
    total_similarity = 0
    for rank, idx in enumerate(top_indices):
        book_info = df.iloc[idx]  # Retrieve book data from DataFrame
        similarity_score = cosine_scores_np[idx]  # Get similarity score
        total_similarity += similarity_score

        print(f"Rank {rank + 1}:")
        print(f"Title: {book_info['Title']}")
        print(f"Author: {book_info['Author']}")
        print(f"Genres: {book_info['Genres']}")
        print(f"Description: {book_info['Description']}")
        print(f"Similarity Score: {similarity_score:.4f}")
        print(f"index: {idx}")
        print("-" * 50)  # Separator for readability

    # Calculate and store the average similarity score for this query
    avg_similarity = total_similarity / 10
    similarity_scores.append(avg_similarity)

    print(f"Average Similarity Score for Query {query_idx + 1}: {avg_similarity:.4f}")
    print("=" * 50)  # Better separator for queries

# Compute overall average similarity score across all queries
overall_avg_similarity = sum(similarity_scores) / len(similarity_scores)

print(f"\n===== Overall Average Similarity Score Across All Queries: {overall_avg_similarity:.4f} =====")
```
